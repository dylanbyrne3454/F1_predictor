```python
import pandas as pd
import fastf1
import time 
```

# Getting the data

## Creating training data

```python
fastf1.Cache.enable_cache("f1_cache")

training_seasons = [2022, 2023, 2024]
all_results = []

for year in training_seasons:
    schedule = fastf1.get_event_schedule(year)

    for index, event in schedule.iterrows():
        if event['EventFormat'] == 'testing':
            continue
        try:
            time.sleep(3)
            session = fastf1.get_session(year, event['RoundNumber'], 'R')
            session.load(laps=False, telemetry=False, weather=True)

            results = session.results[['Abbreviation', 'TeamName', 
                                       'GridPosition', 'Position', 'Points']].copy()
            
            results['Season'] = year
            results['RoundNumber'] = event['RoundNumber']
            results['EventName'] = event['EventName']
            results['Podium'] = (results['Position'] <= 3).astype(int)

            # Weather data
            weather_data = session.weather_data  # no parentheses - it's a property not a method
            results['Rainfall'] = weather_data['Rainfall'].any()
            results['AirTemp'] = weather_data['AirTemp'].median()  # capital T in AirTemp

            # Qualifying session
            session_q = fastf1.get_session(year, event['RoundNumber'], 'Q')
            session_q.load(laps=False, telemetry=False)
            
            quali_results = session_q.results[['Abbreviation', 'Q3']].copy()
            quali_results = quali_results.rename(columns={'Q3': 'QualiTime'})
            quali_results['Season'] = year
            quali_results['RoundNumber'] = event['RoundNumber']

            # Merge quali into results
            results = results.merge(quali_results, on=['Abbreviation', 'Season', 'RoundNumber'], how='left')
            
            all_results.append(results)  # only one append
            print(f"✓ {event['EventName']} {year}")

        except Exception as e:
            print(f"✗ Could not load {event['EventName']}: {e}")

results_df = pd.concat(all_results, ignore_index=True)
results_df.to_csv("data/race_results_22_24.csv", index=False)
print(f"Done — {len(results_df)} rows")
```
